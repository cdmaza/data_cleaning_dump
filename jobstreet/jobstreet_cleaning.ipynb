```python
# 1. company - similarity (consistent letter case, only sdn bhd capital)
df['company'] = df['company'].str.title()

def format_sdn_bhd(company_name):
    if pd.isna(company_name) or re.search("Sdn..*Bhd.", company_name, flags=re.IGNORECASE):  
        return company_name
    company_name = re.sub(r'\b(Sdn)\b', r'\1.', company_name, flags=re.IGNORECASE)
    company_name = re.sub(r'\b(Bhd)\b', r'\1.', company_name, flags=re.IGNORECASE)
    
    return company_name 

df['company'] = df['company'].apply(format_sdn_bhd)
```

```python
# 4. manufacture - change to number
def get_numbers(category_list):
    category_mapping = {}
    for index, category in enumerate(category_list):
        category_mapping[category] = f"{index:02d}"
    return category_mapping

category_list = df["category"].unique()
category_list = sorted(category_list)
category_mapping = get_numbers(category_list)

df['category'] = df['category'].map(category_mapping)
```

```python
# 7. salary - to common format
# create 3 column - minimum, maximum, type_salary, benefit
# common usage word been map
type_month_list= ["per month"," p.m.","monthly","month"]
type_year_list= ["per year","per annum"," p.a."]
type_hour_list= ["per hour"," p.h."]

conversion_rates = {
    "sgd": 3.3162,  
    "$": 4.506,    
    "euro": 4.6822,
    "myr": 1.0
}

def get_type_salary(salary):
    if isinstance(salary, str):

        for keyword in type_month_list:
            if re.search(keyword, salary, flags=re.IGNORECASE):
                return "monthly"
        
        for keyword in type_year_list:
            if re.search(keyword, salary, flags=re.IGNORECASE):
                return "yearly"
        
        for keyword in type_hour_list:
            if re.search(keyword, salary, flags=re.IGNORECASE):
                return "hourly"
    
    return "monthly"

def change_currency(salary, numbers):
    if isinstance(salary, str):
        converted_numbers = []
        currency = "myr"
        
        if re.search(r'\$', salary):
            currency = "$"
        elif re.search(r'euro', salary, flags=re.IGNORECASE):
            currency = "euro"
        elif re.search(r'sgd', salary, flags=re.IGNORECASE):
            currency = "sgd"
        elif re.search(r'myr', salary, flags=re.IGNORECASE):
            currency = "myr"
        
        for num in numbers:
            if 'k' in num.lower():
                num = float(num.lower().replace('k', '')) * 1000
            elif ',' in num:
                num = num.replace(',', '')
            elif '.' in num:
                num = float(num.replace('.', '')) * 1000
            converted_numbers.append(float(num) * conversion_rates.get(currency, 1.0))
        
        return converted_numbers
    
    return []

def extract_salaries(salary):
    if isinstance(salary, str):

        salary = re.sub(r'[^\x00-\x7F]+', ' ', salary)       

        numbers = re.findall(r'\d+(?:,\d+)*(?:\.\d+)?[kK]?', salary)
        
        if not numbers:
            return pd.Series([None, None, None, salary])
        
        extracted_nums = change_currency(salary, numbers)

        min_salary = extracted_nums[0] if len(extracted_nums) > 1 else None
        max_salary = extracted_nums[1] if len(extracted_nums) > 1 else extracted_nums[0]

        return pd.Series([min_salary, max_salary, get_type_salary(salary), None])
    
    return pd.Series([None, None, None, None])


df[['min_salary', 'max_salary', 'type_salary', 'benefit']] = df['salary'].apply(extract_salaries)
```

```python
# 8. date - to common format
df['date'] = pd.to_datetime(df['listingDate']).dt.date
```

```python
# # drop for missing values
df = df.dropna(subset=['min_salary', 'max_salary'], how='all')
```
